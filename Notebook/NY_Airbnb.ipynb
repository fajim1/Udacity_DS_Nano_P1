```python
# Necessary imports
import numpy as np
import pandas as pd
from IPython import display


import os  

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 101)

import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the dataset
df_ny = pd.read_csv("../data/AB_NYC_2019.csv")
df_ny.head()
```

```python
df_ny['neighbourhood_group'].value_counts()[:20]
```

```python
df_manhattan = df_ny[df_ny['neighbourhood_group']=='Manhattan']
df_manhattan.head()
```

```python
df_manhattan.describe()
```

```python
df_manhattan.isnull().sum()
```

```python
from scipy import stats
temp = df_manhattan[['price','number_of_reviews']]
plt.figure(figsize=(10,10))
sns.scatterplot(x='price', y='number_of_reviews', data=temp[(np.abs(stats.zscore(temp)) < 3).all(axis=1)],alpha=0.9)
plt.show()
```

```python
group = df_manhattan[['neighbourhood','price']].groupby('neighbourhood').mean()
group.sort_values(by=['price'],ascending=False)
```

```python
group = df_manhattan[['neighbourhood','number_of_reviews']].groupby('neighbourhood').sum()
group.sort_values(by=['number_of_reviews'],ascending=False)
```

```python
#let's what we can do with our given longtitude and latitude columns

#let's see how scatterplot will come out 

df_manhattan.plot(kind='scatter', x='longitude', y='latitude', c=np.log(df_manhattan['price']),
                         s =np.log(df_manhattan['price'])/5,
                  cmap=plt.get_cmap('YlGnBu'), colorbar=True, alpha=0.4, figsize=(10,10),subplots=True, layout=(1,1))
df_manhattan.plot(kind='scatter', x='longitude', y='latitude', c=np.log(df_manhattan['number_of_reviews']),
                         s =np.log(df_manhattan['number_of_reviews'])/5,
                  cmap=plt.get_cmap('YlGn'), colorbar=True, alpha=0.4, figsize=(10,10),subplots=True, layout=(1,1))
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='longitude', y='latitude',hue='neighbourhood',s=5, data=df_manhattan,alpha=0.9)
plt.show()
```

```python
index = df_manhattan['neighbourhood'].value_counts()[:20].index
```

```python

plt.figure(figsize=(10,10))
sns.scatterplot(x='longitude', y='latitude',hue='neighbourhood',s=5, data=df_manhattan[df_manhattan['neighbourhood'].isin(index)],alpha=0.9)
plt.show()
```
